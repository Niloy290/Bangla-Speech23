```python
data='/content/drive/MyDrive/B_Sp_k23/Sorted_dataset/উত্তেজিত'
```

```python
dir_list=os.listdir(data)

dir_list
```

```python
dataset_df_1=data
text=''
sentence=[]
for dirname, _, filenames in os.walk(dataset_df_1):
    for filename in filenames:
          p=(os.path.join(dirname, filename))
        
          r=sr.Recognizer()
          with sr.AudioFile(p) as source:
              audio=r.record(source)
              try:
               text=r.recognize_google(audio, language='bn_BD')
              except sr.UnknownValueError:
               text = "Google Speech Recognition could not understand audio"
        
          
          #print(text)
          sentence.append(text)
sentence       
```

```python
senText=pd.DataFrame(sentence,columns=['text'])
senText
```

```python
path=[]
emotion=[]

for i in dir_list:
     path.append(data+i)
     emotion.append("Excited")

new_dataset2=pd.DataFrame(path,columns=['path'])
new_dataset2['Text']=senText
new_dataset2['class']=pd.DataFrame(emotion)

new_dataset2
```
